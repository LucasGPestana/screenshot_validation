Existem algumas colunas númericas que apresentam apenas um valor único. Essas colunas não agregarão valor ao algoritmo de aprendizado de máquina e, logo, serão removidas.

```python
# Removendo as colunas numéricas com um único valor

df_numeric = df.select_dtypes("number")

unique_value_numeric_columns = (
    df_numeric.columns.sort_values()[
        df_numeric.nunique().sort_index() == 1
        ]
)

df = df.drop(
    unique_value_numeric_columns,
    axis=1
)

df.shape
```

```python
# Verificando se há valores ausentes
df[df.isnull().any(axis=1)]
```

Também pode ser removido a coluna `nome_arquivo`, uma vez que ela apresenta valores unicos em todos os registros e, por conta disso, não servirá para captura de padrões pelo algoritmo de ml

```python
# Removendo a coluna nome_arquivo
df = df.drop(["nome_arquivo"], axis=1)
df.shape
```

```python
# Verificando a proporção dos valores do atributo alvo
df["rotulo"].value_counts(normalize=True)
```

```python
# Verificando se há inconsistências entre os registros duplicados
df.groupby(
    df.columns.drop("rotulo").tolist()
).filter(lambda x: x["rotulo"].nunique() > 1)
```

Percebe-se que há inconsistências entre registros que possuem os mesmos valores nas demais colunas, mas apresentam valores diferentes na coluna alvo.

Apesar de boa parte ser da classe minoritária, não é interessante deixá-los na base, pois o algoritmo poderá rotular erroneamente registros com valores próximos a esses.

```python
# Removendo os registros duplicados e inconsistentes

incosistent_indexes = (
    df.groupby(
        df.columns.drop("rotulo").tolist()
    ).filter(
        lambda x: x["rotulo"].nunique() > 1
        ).index
)

df = df[~(df.index.isin(incosistent_indexes))].reset_index(drop=True)
df["rotulo"].value_counts(normalize=True)
```

```python
# Salvando as modificações
df.to_parquet(CLEANED_DATA_FILEPATH, index=False)
```
